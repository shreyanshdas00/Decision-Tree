```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sb
from DecisionTreeClassifier import DecisionTreeClassifier
```

```
iris_dataset = load_iris()
class_labels = iris_dataset.target
class_label_names = iris_dataset.target_names
iris_dataset = pd.DataFrame(list(iris_dataset.data), columns = iris_dataset.feature_names)
iris_dataset['Label'] = class_labels
iris_dataset['Species'] = [class_label_names[class_label] for class_label in class_labels]
```

```
iris_dataset
```

```
fig, axes = plt.subplots(1, 2,  sharex=True, figsize=(14,7))
sb.scatterplot(data = iris_dataset, x = 'sepal length (cm)', y = 'petal length (cm)', hue = 'Species', ax = axes[0])
sb.scatterplot(data = iris_dataset, x = 'sepal width (cm)', y = 'petal width (cm)', hue = 'Species', ax = axes[1])
plt.show()
fig, axes = plt.subplots(1, 2,  sharex=True, figsize=(14,7))
sb.scatterplot(data = iris_dataset, x = 'petal length (cm)', y = 'sepal width (cm)', hue = 'Species', ax = axes[0])
sb.scatterplot(data = iris_dataset, x = 'sepal length (cm)', y = 'petal width (cm)', hue = 'Species', ax = axes[1])
plt.show()
fig, axes = plt.subplots(1, 2,  sharex=True, figsize=(14,7))
sb.scatterplot(data = iris_dataset, x = 'sepal length (cm)', y = 'sepal width (cm)', hue = 'Species', ax = axes[0])
sb.scatterplot(data = iris_dataset, x = 'petal length (cm)', y = 'petal width (cm)', hue = 'Species', ax = axes[1])
plt.show()
```

```
training_set_data, test_set_data, training_set_labels, test_set_labels = train_test_split(iris_dataset.drop(columns = ['Label', 'Species']), iris_dataset['Label'], test_size=0.2)
```

```
print('Training dataset size:', len(training_set_data))
print('Testing dataset size:', len(test_set_data))
```

```
training_set_data.sort_index()
```

```
test_set_data.sort_index()
```

```
tree = DecisionTreeClassifier()
```

```
tree.fit(training_set_data, training_set_labels, test_set_data, test_set_labels)
```

Predictions are of the form [(label_1, probability(label_1)), ...(label_n, probability(label_n))

```
final_predictions_train = tree.predict(training_set_data)
final_predictions_train
```

```
final_predictions_test = tree.predict(test_set_data)
final_predictions_test
```

```
print("Accuracy on training set:", tree.evaluate_accuracy(training_set_data, training_set_labels))
print("Accuracy on test set:", tree.evaluate_accuracy(test_set_data, test_set_labels))
```

Limiting the depth of the tree

```
depth_limited_tree = DecisionTreeClassifier(max_depth = 2)
```

```
depth_limited_tree.fit(training_set_data, training_set_labels, test_set_data, test_set_labels)
```

```
leaf_limited_tree = DecisionTreeClassifier(max_leaves = 3)
```

```
leaf_limited_tree.fit(training_set_data, training_set_labels, test_set_data, test_set_labels)
```

Tree with specified min_child_weight required for spliting

```
controlled_split_tree = DecisionTreeClassifier(min_child_weight = 50)
```

```
controlled_split_tree.fit(training_set_data, training_set_labels, test_set_data, test_set_labels)
```
