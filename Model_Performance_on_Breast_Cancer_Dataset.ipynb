```
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sb
from DecisionTreeClassifier import DecisionTreeClassifier
```

```
breast_cancer_dataset = load_breast_cancer()
class_labels = breast_cancer_dataset.target
class_label_names = breast_cancer_dataset.target_names
breast_cancer_dataset = pd.DataFrame(list(breast_cancer_dataset.data), columns = breast_cancer_dataset.feature_names)
breast_cancer_dataset['Label'] = class_labels
breast_cancer_dataset['Cancer Type'] = [class_label_names[class_label] for class_label in class_labels]
```

```
breast_cancer_dataset
```

```
features_mean = breast_cancer_dataset.drop(columns = ['Label', 'Cancer Type'])
plt.figure(figsize=(30,30))
heat = sb.heatmap(breast_cancer_dataset.drop(columns = ['Label', 'Cancer Type']).corr(), vmax=1, square=True, annot=True)
plt.title('Data Heatmap')
```

```
pca = PCA(n_components = 2) #Binary Classifier
pca = pca.fit_transform(breast_cancer_dataset.drop(columns = ['Label', 'Cancer Type']))
plt.figure(figsize = (6,6))
plt.scatter(pca[:, 0], pca[:,1], c = breast_cancer_dataset['Label'], cmap = "winter", edgecolor = "Black", alpha=0.35)
plt.colorbar()
plt.title('PCA Scatter Plot')
```

```
tsne = TSNE(verbose = 0, perplexity = 80, n_iter = 8000)
tsne = tsne.fit_transform(breast_cancer_dataset.drop(columns = ['Label', 'Cancer Type']))
plt.figure(figsize = (6,6))
plt.scatter(tsne[:, 0], tsne[:, 1],  c = breast_cancer_dataset['Label'], cmap = "winter", edgecolor = "Black", alpha=0.35)
plt.colorbar()
plt.title('t-SNE Scatter Plot')
```

```
training_set_data, test_set_data, training_set_labels, test_set_labels = train_test_split(breast_cancer_dataset.drop(columns = ['Label', 'Cancer Type']), breast_cancer_dataset['Label'], test_size = 0.20)
```

```
print('Training set size:', len(training_set_data))
print('Test set size:', len(test_set_data))
```

```
training_set_data.sort_index()
```

```
test_set_data.sort_index()
```

```
decision_tree_classifier = DecisionTreeClassifier()
```

```
decision_tree_classifier.fit(training_set_data, training_set_labels, test_set_data, test_set_labels)
```

```
final_training_set_predictions = decision_tree_classifier.predict(training_set_data)
final_training_set_predictions
```

```
final_test_set_predictions = decision_tree_classifier.predict(test_set_data)
final_test_set_predictions
```

```
print('Accuracy on training data:', decision_tree_classifier.evaluate_accuracy(training_set_data, training_set_labels))
print('Accuracy on test data:', decision_tree_classifier.evaluate_accuracy(test_set_data, test_set_labels))
```
